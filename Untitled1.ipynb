## Sınıflandırma

#### 1. K Means 

```python
#sınıflandırma için kullanılacak sütunları seçme
features=['longitude', 'latitude', 'median_income']
select_df = df[features]
select_df.columns
```

```python
#veri standardizasyonu
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(select_df)
x[:5]
```

```python
#scikit leanr kütüphanesinden kmeans i çağırma
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4) 
model= kmeans.fit(x) #modeli eğitme
print(model)
```

```python
centers = model.cluster_centers_
centers[:5]
```

```python
#merkez değişimlerini gözlemleyebilmek adına tasarlanan fonksiyonlar
from itertools import cycle, islice
from pandas.plotting import parallel_coordinates
def pd_centers(featuresUsed, centers):
    colNames = list(featuresUsed)
    colNames.append('prediction')
    z = [np.append(A, index) for index, A in enumerate(centers)]
    p= pd.DataFrame(z, columns=colNames)
    p['prediction'] = p['prediction'].astype(int)
    return p
```

```python
# merkez değişimini görselleştirme
def parallel_plot(df):
    my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(df)))
    plt.figure(figsize=(15,8)).gca().axes.set_ylim([-3,+3])
    parallel_coordinates(df, 'prediction', color = my_colors, marker='o')
```

```python
p = pd_centers(features, centers)
p
```

```python
parallel_plot(p)
```

```python
#standardizasyonu yapılan verilere göre kümeleme işlemi
df['econ_region'] = kmeans.fit_predict(x)
df['econ_region'] = df['econ_region'].astype("category")
df.head()
```

## TENSORFLOW

```python
import tensorflow as tf
```

```python
#özniteliklerin belirlenmesi
A=pd.DataFrame(columns=['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population',
                        'households','median_income','ocean_proximity'],data=df)
B=pd.DataFrame(columns=['median_house_value'],data=df)
```

```python
#kategorik verinin düzenlenmesi
A = pd.get_dummies(data = A, columns = ['ocean_proximity'] , prefix = ['ocean_proximity'] , drop_first = True)
A
```

Makine öğrenimi ve derin öğrenmenin temeli matematiktir. Bu sebeple elimizde sayısal olmayan veriler varsa bunları sayısal olarak ifade etmeliyiz. Aksi takdirde verilerimiz çalışmalarımızda kullanılamayacaktır.Dummies ile verilerimiz one-hot-encoding şeklinde güncellenir.

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size = 0.3, random_state = 0)
```

```python
#veri standardizasyonu
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
A_train = scaler.fit_transform(A_train)#modeli eğitip çıktı alma(fit_transform)
A_test = scaler.transform(A_test)#modelden sadece çıktı alma(transform)
```

```python
A_train
```

### ANN

#### ANN Çalışma 1

```python
# gerekli kütüphanelerin çağırılması
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras import optimizers
# model oluşturma 
# Sequential model sıralı katmanlardan oluşan bir modeldir.
modelA = Sequential()

#Input Layer
modelA.add(Dense(A.shape[1], activation='relu', input_dim = A.shape[1]))
#kernel initializer ağırlıkları başlatmak için kullanılan komuttur.
#activation function için ise relu tercih edilmiştir. reluda girdi 0dan büyükse olduğu gibi kabul edilir değilse 0
#Hidden Layer
modelA.add(Dense(512,kernel_initializer='normal', activation='relu'))
modelA.add(Dense(512,kernel_initializer='normal', activation='relu'))
modelA.add(Dense(256,kernel_initializer='normal', activation='relu'))
modelA.add(Dense(128,kernel_initializer='normal', activation='relu'))
modelA.add(Dense(64,kernel_initializer='normal', activation='relu'))
modelA.add(Dense(32,kernel_initializer='normal', activation='relu'))


#Output Layer
modelA.add(Dense(1,kernel_initializer='normal', activation = 'relu'))
```

1 Giriş katmanı, 6 ara katman ve 1 çıkış katmanı olmak üzere 8 katmanlı bir yapay sinir ağıdır.

```python
modelA.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
modelA.summary()
```

Eğitimde değerlendirme olarak 

    -[MSE] Ortalama Kare Hata  (Mean Squared Error)
    -[MAE] Ortalama Mutlak Hata (Mean Absolute Error)
kullandım.
Optimizer olarak doğruluğu en yüksek olan adam yöntemini seçtim.

```python
# batch size -> Tek bir grupta bulunan toplam eğitim örneği sayısı
# epochs -> işlemin tekrar etme sayısını
historyA = modelA.fit(A_train, B_train.to_numpy(), batch_size = 32, epochs = 10, validation_split=0.1)
```

```python
y_pred = modelA.predict(A_test)
y_pred
```

```python
y_test
```

```python
# görünmeyen test verilerindeki doğruluğun kaybın kontrolü
modelA.evaluate(A_test, B_test)
```

```python
plt.plot(historyA.history['loss'])
plt.plot(historyA.history['val_loss'])
plt.title('LOSS')
plt.ylabel('LOSS')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
```

Loss değerinin 0'a yakın olması istenmektedir.

```python
# Plot training & validation accuracy values
plt.figure(figsize=(14,3))
plt.subplot(1, 2, 1)
plt.plot(historyA.history['mse'])
plt.plot(historyA.history['val_mse'])
plt.title('Model mse')
plt.ylabel('MSE')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(historyA.history['mae'])
plt.plot(historyA.history['val_mae'])
plt.title('Model mae')
plt.ylabel('MAE')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

MSE, tahmin edilen sonuçların gerçek sayıdan ne kadar farklı olduğuna dair mutlak bir sayı verir. Bu sebeple değerinin sıfıra yakın olması beklenir. MAE ise mutlak hata değerinin toplamını alır. Bu sebeple yine düşük bir değer olması beklenir.

#### ANN Çalışma 2

```python
#Üstteki ANN çalışmamızı aynı değerler ile ancak daha az katman ile deneyelim ve sonuca etkisini yorumlayalım
modelB = Sequential()

#Input Layer
modelB.add(Dense(A.shape[1], activation='relu', input_dim = A.shape[1]))

#Hidden Layer
modelB.add(Dense(512,kernel_initializer='normal', activation='relu'))
modelB.add(Dense(512,kernel_initializer='normal', activation='relu'))
modelB.add(Dense(256,kernel_initializer='normal', activation='relu'))

#Output Layer
modelB.add(Dense(1,kernel_initializer='normal', activation = 'relu'))

modelB.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
modelB.summary()
```

1 Giriş katmanı, 3 ara katman ve 1 çıkış katmanı olmak üzere 8 katmanlı bir yapay sinir ağıdır.

```python
historyB=modelB.fit(A_train,B_train, batch_size=32,epochs=10,validation_split=0.1)
```

```python
plt.plot(historyB.history['loss'])
plt.plot(historyB.history['val_loss'])
plt.title('LOSS')
plt.ylabel('LOSS')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
```

Daha az katman aldığımız sonucu kötü etkilemiştir. Bu sebeple gizli katman sayısının çok olması olumlu bir etkiye sahiptir.

```python
plt.figure(figsize=(14,3))
plt.subplot(1, 2, 1)
plt.plot(historyB.history['mse'])
plt.plot(historyB.history['val_mse'])
plt.title('Model mse')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(historyB.history['mae'])
plt.plot(historyB.history['val_mae'])
plt.title('Model mae')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
